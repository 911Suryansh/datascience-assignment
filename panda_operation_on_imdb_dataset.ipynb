### **step 3 :  analyze the data**

```python
top_movies = df.sort_values(by='popularity', ascending=False).head(10)
top_movies[['title', 'popularity']]
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df['Year'] = df['release_date'].dt.year
df['Year'].value_counts().sort_index()
```

```python
df.groupby('Year')['popularity'].mean()
```

```python
import ast

def get_genre_names(genres_str):
    if pd.isna(genres_str) or genres_str == '':
        return []
    try:
        genres_list = ast.literal_eval(genres_str)
        if isinstance(genres_list, list):
            return [g['name'] for g in genres_list if isinstance(g, dict) and 'name' in g]
        return []
    except (ValueError, SyntaxError):
        return []

df['parsed_genres'] = df['genres'].apply(get_genre_names)
df_genre = df.explode('parsed_genres')
df_genre = df_genre.rename(columns={'parsed_genres': 'genres'})
```

```python
df_genre['genres'].value_counts()
```

```python
df_genre.groupby('genres')['popularity'].mean().sort_values(ascending=False)
```

### **visualizing the data**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.hist(df['popularity'], bins=20)
plt.xlabel("Popularity")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Popularity")
plt.show()
```

```python
movies_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', linestyle='-', color='green')

plt.title('Number of Movies Released Per Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.grid(True)
plt.show()
```
